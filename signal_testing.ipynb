```python
import numpy as np

import matplotlib.pyplot as plt

# Parameters
length = 100000
fs = 2e6  # Sample rate (at least 2 * max_freq to satisfy Nyquist)
min_freq = 1.0  # Hz
max_freq = 1e6  # 1 MHz
min_prob = 0.02  # Probability at max_freq
max_prob = 0.8  # Probability at min_freq

# Generate time array
t = np.arange(length) / fs

# Generate frequencies with probabilities that decrease linearly from 80% to 2%
num_freq_components = 1000  # Number of frequency components to consider
freqs = np.linspace(min_freq, max_freq, num_freq_components)
probs = np.linspace(max_prob, min_prob, num_freq_components)

# Initialize signal
signal = np.zeros(length)

# Generate composite signal
for i, (freq, prob) in enumerate(zip(freqs, probs)):
    if np.random.random() < prob:  # Include this frequency with probability 'prob'
        phase = 2 * np.pi * np.random.random()  # Random phase
        amplitude = np.random.random()  # Random amplitude
        signal += amplitude * np.sin(2 * np.pi * freq * t + phase)

# Normalize signal
signal = signal / np.max(np.abs(signal))

# Plot a section of the signal
plt.figure(figsize=(12, 6))
plt.plot(t[:1000], signal[:1000])
# plt.plot(t, signal)
plt.title('Signal with Random Frequency Components')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)
plt.tight_layout()
```

```python
fs_new = 400e3  #  400 kHz
downsample_factor = int(fs / fs_new)  # 2MHz/400kHz = 5
t_downsampled = t[::downsample_factor]
signal_downsampled = signal[::downsample_factor]

# 
plt.figure(figsize=(12, 6))
plt.plot(t_downsampled[:200], signal_downsampled[:200])
plt.title('Down Sampling (400kHz)')
plt.xlabel('Time (s)')
plt.ylabel('Voltage')
plt.grid(True)
plt.tight_layout()
```
